```python
#Use scipy.linalg.solve() to calculate energy from solar panels and batteries
x = solar
y = battery
B = np.array([D1, D2])
x, y = solve(A, B)

print(f"Day {day} -> Solar panels: {x:.2f}, Batteries: {y:.2f}")
solar.append(x)
battery.append(y)
```

```python
#Repeat for 7 days, storing results in a NumPy array
np.random.seed(42)  # reproducibility
D1_days = np.random.randint(10, 50, size=7)
D2_days = np.random.randint(5, 40, size=7)

solar = []
battery = []

print("Day-wise Simulation of Solar and Battery Energy:\n")
for day, (D1, D2) in enumerate(zip(D1_days, D2_days), start=1):
    # Solve linear system for x (solar) and y (batteries)
    x, y = solve(A, [D1, D2])
    solar.append(x)
    battery.append(y)
    print(f"Day {day}: D1={D1}, D2={D2} -> Solar={x:.2f}, Battery={y:.2f}")

```

```python
#Use statistics to compute mean, variance, and standard deviation of solar vs
#battery usage
if len(solar) > 1:
    print("\nSolar Panels:")
    print("Mean:", stats.mean(solar))
    print("Variance:", stats.variance(solar))
    print("Std Dev:", stats.stdev(solar))

    print("\nBatteries:")
    print("Mean:", stats.mean(battery))
    print("Variance:", stats.variance(battery))
    print("Std Dev:", stats.stdev(battery))
else:
    print("Not enough data to compute variance or std dev.")
```

```python
#Plot daily solar vs battery usage.
days = range(1, 8)
plt.plot(days, solar, 'bo-', label='Solar Panels')
plt.plot(days, battery, 'ro-', label='Batteries')
plt.xlabel("Day")
plt.ylabel("Energy Output")
plt.title("Daily Solar vs Battery Usage")
plt.legend()
plt.grid(True)
plt.show()
```

```python
#####   Lake Victoria Fish Export Risk Model
import numpy as np
import statistics as stats
import matplotlib.pyplot as plt
#Generate 15 Fibonacci numbers to simulate fish stock growth
fish_stock = [0, 1]
for i in range(2, 15):
    fish_stock.append(fish_stock[-1] + fish_stock[-2])
fish_stock = np.array(fish_stock)
print("Fish stock (kg):", fish_stock)
```

```python
# Simulate fish prices in UGX per kg, e.g. [12000, 12500, 11800, 13000, 12800,...].

fish_price = np.array([12000, 12500, 11800, 13000, 12800, 13200, 12900,
                       13500, 14000, 13800, 14200, 14500, 14700, 15000, 15500])
print("Fish prices (UGX/kg):", fish_price)
```

```python
# Multiply stock  price to estimate daily revenue
revenue = fish_stock * fish_price
print("Daily revenue (UGX):", revenue)
```

```python
# Compute mean, median, variance of revenue using statistics
mean_rev = stats.mean(revenue)
median_rev = stats.median(revenue)
var_rev = stats.variance(revenue)

print("\nRevenue Statistics:")
print("Mean:", mean_rev)
print("Median:", median_rev)
print("Variance:", var_rev)
```

```python
# Print 'High risk' if variance > 50,000, else 'Low risk'.
if var_rev > 50000:
    print("Risk Level: High risk ")
else:
    print("Risk Level: Low risk ")
```

```python
# Plot fish stock vs revenue trend
plt.figure(figsize=(8,5))
plt.plot(fish_stock, revenue, 'bo-', label='Revenue Trend')
plt.xlabel("Fish Stock (kg)")
plt.ylabel("Revenue (UGX)")
plt.title("Fish Stock vs Revenue Trend")
plt.legend()
plt.grid(True)
plt.show()
```

```python
### Weather & Agriculture Analysis
import numpy as np
import math
import matplotlib.pyplot as plt
#Store rainfall data in a NumPy array
Kampala = np.array([120, 140, 180, 200, 220, 180, 90, 70, 60, 100, 110, 130])
```

```python
#Compute mean rainfall and classify months
mean_rain = np.mean(Kampala)
print("Mean Rainfall (Kampala):", mean_rain)

for i, rain in enumerate(Kampala, start=1):
    if rain > 150:
        print(f"Month {i}: Good for maize  ({rain} mm)")
    else:
        print(f"Month {i}: Drought risk  ({rain} mm)")
```

```python
#Simulate Gulu rainfall (different values)
Gulu = np.array([100, 120, 160, 190, 210, 160, 80, 60, 70, 90, 100, 120])
print("\nGulu Rainfall Data:", Gulu)
```

```python
#Compute cosine similarity of Kampala vs Gulu rainfall
dot_product = np.dot(Kampala, Gulu)
magnitude_k = math.sqrt(np.dot(Kampala, Kampala))
magnitude_g = math.sqrt(np.dot(Gulu, Gulu))
cos_similarity = dot_product / (magnitude_k * magnitude_g)
print("\nCosine Similarity between Kampala and Gulu:", round(cos_similarity, 3))
```

```python
# Plot both regions' rainfall
months = np.arange(1, 13)
plt.plot(months, Kampala, 'bo-', label='Kampala Rainfall')
plt.plot(months, Gulu, 'ro-', label='Gulu Rainfall')
plt.xlabel('Month')
plt.ylabel('Rainfall (mm)')
plt.title('Monthly Rainfall Comparison: Kampala vs Gulu')
plt.legend()
plt.grid(True)
plt.show()
```
